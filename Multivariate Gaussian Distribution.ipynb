### Log likelihood for MVN

$$
\begin{aligned}
l\left(\mu, \boldsymbol{\Sigma} \mid \mathbf{x}^{(\mathbf{i})}\right) &=\log \prod_{i=0}^{m-1} f_{\mathbf{X}^{(\mathrm{i})}}\left(\mathbf{x}^{(\mathbf{i})} \mid \mu, \mathbf{\Sigma}\right) \\
&=\log \prod_{i=0}^{m-1} \frac{1}{(2 \pi)^{p / 2}|\Sigma|^{1 / 2}} \exp \left(-\frac{1}{2}\left(\mathbf{x}^{(\mathbf{i})}-\mu\right)^{\mathbf{T}} \mathbf{\Sigma}^{-\mathbf{1}}\left(\mathbf{x}^{(\mathbf{i})}-\mu\right)\right) \\
&=\sum_{i=0}^{m-1}\left(-\frac{p}{2} \log (2 \pi)-\frac{1}{2} \log |\Sigma|-\frac{1}{2}\left(\mathbf{x}^{(\mathbf{i})}-\mu\right)^{\mathbf{T}} \mathbf{\Sigma}^{-\mathbf{1}}\left(\mathbf{x}^{(\mathbf{i})}-\mu\right)\right) \\
l(\mu, \Sigma ;) &=-\frac{m p}{2} \log (2 \pi)-\frac{m}{2} \log |\Sigma|-\frac{1}{2} \sum_{i=0}^{m-1}\left(\mathbf{x}^{(\mathbf{i})}-\mu\right)^{\mathbf{T}} \mathbf{\Sigma}^{-\mathbf{1}}\left(\mathbf{x}^{(\mathbf{i})}-\mu\right)
\end{aligned}
$$

### MLE for MVN
$$
\begin{equation}
\begin{aligned}
&\hat{\mu}=\frac{1}{m} \sum_{i=1}^{m} \mathbf{x}^{(\mathbf{i})}=\overline{\mathbf{x}} \\
&\hat{\Sigma}=\frac{1}{m} \sum_{i=1}^{m}\left(\mathbf{x}^{(\mathbf{i})}-\hat{\mu}\right)\left(\mathbf{x}^{(\mathbf{i})}-\hat{\mu}\right)^{T}
\end{aligned}
\end{equation}
